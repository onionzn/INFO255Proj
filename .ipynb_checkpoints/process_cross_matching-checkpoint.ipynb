```python
import pandas as pd
import numpy as np

df_reviews_with_first_name = pd.read_csv('./data/reviews_with_first_name.csv')
```

```python
df_reviews_with_first_name
```

```python
df_tinder_users_no_dup = pd.read_csv('./data/tinder_users_no_dup.csv')
```

```python
df_tinder_users_no_dup
```

```python
merged_df = pd.merge(df_reviews_with_first_name, df_tinder_users_no_dup, left_on='user_first_name', right_on='name')
merged_df.to_csv("merged.csv")
```

```python
matched = merged_df[merged_df['image_match']>0.4]
matched
```

```python
highest = matched[matched['image_match']>0.5]
highest
```
